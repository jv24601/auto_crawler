```python
import pandas as pd
import numpy as np 
from sklearn.decomposition import PCA
import matplotlib.pyplot as plot
```

```python
jazz_df = pd.read_csv("jazz.csv")
N = 19
K = 8
```

```python
jazz_df.head()
```

```python
pca = PCA(n_components=N)
normalized_jazz_df = (jazz_df - jazz_df.mean())/jazz_df.std()
```

```python
pca.fit(normalized_jazz_df)
```

```python
for i in range(N):
    print(list(normalized_jazz_df.columns[np.argsort(pca.components_[i])[-K:]])[::-1])
```
